```python
import trafilatura
homepage = "https://www.theguardian.com/"
downloaded = trafilatura.fetch_url(homepage)
downloaded
```

```python
trafilatura.extract(downloaded, output_format="xml")
```

```python
from trafilatura import feeds
feedurls = feeds.find_feed_urls(homepage)
len(feedurls), feedurls
```

```python
feeds.find_feed_urls("https://www.theguardian.com/international/rss")
```

```python
targets = [u for u in feedurls if "science" in u]
targets
```

```python
from trafilatura import fetch_url, extract
texts = [extract(fetch_url(u)) for u in targets]
texts
```
